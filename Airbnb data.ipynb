# Data Audit

```python
# Count missing values and show as percentage
# Count how many values are missing in each column
missing_count = df.isnull().sum()

# Calculate what percentage of the column is missing
missing_percent = (df.isnull().sum() / len(df) * 100).round(2)

# Combine into one table
missing = pd.DataFrame({
    'missing_count': missing_count,
    'missing_percent': missing_percent
})

# Only show columns that actually have missing values
missing = missing[missing['missing_count'] > 0]
missing = missing.sort_values('missing_percent', ascending=False)

print(missing)
```

```python
# How many listings have a price of zero?
zero_price = df[df['log_price'] == 0]

print("Listings with zero price:", len(zero_price))
print(zero_price[['log_price', 'property_type', 'room_type']].head())
```

```python
# How many listings have zero bathrooms?
zero_bath = df[df['bathrooms'] == 0]

print("Listings with 0 bathrooms:", len(zero_bath))
print(zero_bath[['property_type', 'room_type', 'bathrooms', 'bedrooms']].head(10))
```

# Data Cleaning

```python
# Make a copy
df_clean = df.copy()
```

```python
# These columns are either useless IDs or mostly empty
df_clean = df_clean.drop(columns=['thumbnail_url', 'id'], errors='ignore')
```

```python
# How many listings have a zero price?
zero_price_count = len(df_clean[df_clean['log_price'] == 0])
print("Zero price listings found:", zero_price_count)

# Remove them — $0/night is clearly bad data
df_clean = df_clean[df_clean['log_price'] > 0]

print("Shape after removing zero prices:", df_clean.shape)
```

```python
# Check what the median values are first
print("Bathrooms median:", df_clean['bathrooms'].median())
print("Bedrooms median:", df_clean['bedrooms'].median())
print("Beds median:", df_clean['beds'].median())

# Fill missing values with the median
# We use median because a few huge mansions would pull the average too high
df_clean['bathrooms'] = df_clean['bathrooms'].fillna(df_clean['bathrooms'].median())
df_clean['bedrooms'] = df_clean['bedrooms'].fillna(df_clean['bedrooms'].median())
df_clean['beds'] = df_clean['beds'].fillna(df_clean['beds'].median())

print("Done! Missing values filled.")
```

```python
# log_price is the price after a math transformation called log
# np.exp() reverses that transformation to give us real dollar prices

df_clean['price'] = np.exp(df_clean['log_price']).round(2)

print("Lowest price:  $", df_clean['price'].min())
print("Highest price: $", df_clean['price'].max())
print("Average price: $", round(df_clean['price'].mean(), 2))
```

```python
# How much did we clean up?
print("Original rows:", df.shape[0])
print("Cleaned rows: ", df_clean.shape[0])
print("Rows removed: ", df.shape[0] - df_clean.shape[0])

print("")
print("Any missing values left?")
print(df_clean.isnull().sum()[df_clean.isnull().sum() > 0])
```

# Add Visualizations

```python
import matplotlib.pyplot as plt
```

# What types of properties dominate Airbnb?

```python
property_counts = df_clean['property_type'].value_counts()

print("Top 10 property types:")
print(property_counts.head(10))
```

```python
# Get top 10 property types
top_properties = df_clean['property_type'].value_counts().head(10)

plt.figure(figsize=(10, 5))
plt.bar(top_properties.index, top_properties.values, color='steelblue')
plt.title('Top 10 Property Types on Airbnb')
plt.xlabel('Property Type')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Which room type is most common and most expensive?

```python
print("Room type counts:")
print(df_clean['room_type'].value_counts())

print("")

# groupby to find average price per room type
room_price = df_clean.groupby('room_type')['price'].mean().round(2)
room_price = room_price.sort_values(ascending=False)

print("Average price by room type:")
print(room_price)
```

```python
room_price = df_clean.groupby('room_type')['price'].mean().round(2)
room_price = room_price.sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(room_price.index, room_price.values, color=['steelblue', 'orange', 'green'])
plt.title('Average Price by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Average Price ($)')
plt.tight_layout()
plt.show()
```

# Which cities are the most expensive?

```python
# groupby city and find average price
city_price = df_clean.groupby('city')['price'].mean().round(2)
city_price = city_price.sort_values(ascending=False)

print("Average price by city:")
print(city_price)
```

```python
city_price = df_clean.groupby('city')['price'].mean().round(2)
city_price = city_price.sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(city_price.index, city_price.values, color='coral')
plt.title('Average Airbnb Price by City')
plt.xlabel('City')
plt.ylabel('Average Price ($)')
plt.tight_layout()
plt.show()
```

# Do more bedrooms mean higher price?

```python
# groupby number of bedrooms and find average price
bedroom_price = df_clean.groupby('bedrooms')['price'].mean().round(2)
bedroom_price = bedroom_price.sort_values(ascending=False)

print("Average price by number of bedrooms:")
print(bedroom_price.head(10))
```

```python
bedroom_price = df_clean.groupby('bedrooms')['price'].mean().round(2)
bedroom_price = bedroom_price.sort_values(ascending=True)

# Only show up to 10 bedrooms to avoid outliers
bedroom_price = bedroom_price[bedroom_price.index <= 10]

plt.figure(figsize=(10, 5))
plt.plot(bedroom_price.index, bedroom_price.values, 
         marker='o', color='steelblue', linewidth=2, markersize=8)
plt.title('Average Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price ($)')
plt.tight_layout()
plt.show()
```

# Does having a strict cancellation policy affect price?

```python
# What cancellation policies exist?
print("Cancellation policy counts:")
print(df_clean['cancellation_policy'].value_counts())

print("")

# Does stricter policy = higher price?
cancel_price = df_clean.groupby('cancellation_policy')['price'].mean().round(2)
cancel_price = cancel_price.sort_values(ascending=False)

print("Average price by cancellation policy:")
print(cancel_price)
```

```python
# Remove the super_strict outliers so the chart is readable
cancel_price = df_clean.groupby('cancellation_policy')['price'].mean().round(2)
cancel_price = cancel_price[cancel_price.index.isin(['flexible', 'moderate', 'strict'])]
cancel_price = cancel_price.sort_values(ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(cancel_price.index, cancel_price.values, color='mediumpurple')
plt.title('Average Price by Cancellation Policy')
plt.xlabel('Cancellation Policy')
plt.ylabel('Average Price ($)')
plt.tight_layout()
plt.show()
```
